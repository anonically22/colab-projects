```python
# ============================================================
# FINAL COLAB HISTOPATHOLOGY PIPELINE
# EfficientNet-B0 + SE Attention
# 5-Fold CV on Train
# Final Evaluation on Test
# AMP Safe for Colab
# ============================================================

import os, warnings, random
warnings.filterwarnings("ignore")

import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from torchvision.io import read_image

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score, confusion_matrix, roc_auc_score
from sklearn.preprocessing import label_binarize
import pandas as pd

# ---------------- SETTINGS ----------------
BASE = "/content"
TRAIN_DIR = os.path.join(BASE, "Train_Images")
TEST_DIR  = os.path.join(BASE, "Test_Images")

USE_AUG = True
FREEZE_BACKBONE = False
USE_SE = True

BATCH_SIZE = 64
EPOCHS = 10
NUM_WORKERS = 2

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cudnn.benchmark = True

print("Device:", device)

# ---------------- SE BLOCK ----------------
class SEBlock(nn.Module):
    def __init__(self, ch, r=16):
        super().__init__()
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(ch, ch//r),
            nn.ReLU(),
            nn.Linear(ch//r, ch),
            nn.Sigmoid()
        )
    def forward(self,x):
        b,c,_,_ = x.shape
        y = self.pool(x).view(b,c)
        y = self.fc(y).view(b,c,1,1)
        return x*y

# ---------------- MODEL ----------------
class EfficientNetSE(nn.Module):
    def __init__(self,num_classes):
        super().__init__()
        self.backbone = models.efficientnet_b0(weights="IMAGENET1K_V1")
        self.se = SEBlock(1280) if USE_SE else nn.Identity()
        self.backbone.classifier[1] = nn.Linear(1280,num_classes)

    def forward(self,x):
        x = self.backbone.features(x)
        x = self.se(x)
        x = self.backbone.avgpool(x)
        x = torch.flatten(x,1)
        x = self.backbone.classifier(x)
        return x

# ---------------- COLLECT IMAGES ----------------
def collect_images(base):
    files, labels = [], []
    for cls in sorted(os.listdir(base)):
        p = os.path.join(base, cls)
        if os.path.isdir(p):
            imgs = glob(os.path.join(p,"*"))
            files.extend(imgs)
            labels.extend([cls]*len(imgs))
    return np.array(files), np.array(labels)

train_files, train_labels_str = collect_images(TRAIN_DIR)
test_files, test_labels_str   = collect_images(TEST_DIR)

classes = sorted(np.unique(train_labels_str))
cls_to_idx = {c:i for i,c in enumerate(classes)}

train_labels = np.array([cls_to_idx[i] for i in train_labels_str])
test_labels  = np.array([cls_to_idx[i] for i in test_labels_str])

print("\nClasses:", classes)
print("Train Images:", len(train_files))
print("Test Images:", len(test_files))

# ---------------- DATASET ----------------
class HistoDataset(Dataset):
    def __init__(self, files, labels, tf):
        self.files = files
        self.labels = labels
        self.tf = tf

    def __len__(self): return len(self.files)

    def __getitem__(self,i):
        img = read_image(self.files[i]).float()/255.0
        img = self.tf(img)
        return img, self.labels[i]

train_tf = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(0.2,0.2,0.2)
]) if USE_AUG else transforms.Compose([
    transforms.Resize((224,224))
])

val_tf = transforms.Compose([
    transforms.Resize((224,224))
])

# ---------------- 5-FOLD CV ----------------
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

fold_acc = []

for fold,(tr,va) in enumerate(skf.split(train_files,train_labels),1):

    print(f"\n{'='*50}\nFOLD {fold}\n{'='*50}")

    train_ds = HistoDataset(train_files[tr],train_labels[tr],train_tf)
    val_ds   = HistoDataset(train_files[va],train_labels[va],val_tf)

    train_loader = DataLoader(train_ds,BATCH_SIZE,True,
                              num_workers=NUM_WORKERS,pin_memory=True)
    val_loader   = DataLoader(val_ds,BATCH_SIZE,False,
                              num_workers=NUM_WORKERS,pin_memory=True)

    model = EfficientNetSE(len(classes)).to(device)

    if FREEZE_BACKBONE:
        for p in model.backbone.features.parameters():
            p.requires_grad=False

    optimizer = optim.AdamW(model.parameters(),lr=3e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer,EPOCHS)
    scaler = torch.cuda.amp.GradScaler(enabled=(device.type=="cuda"))
    criterion = nn.CrossEntropyLoss()

    best = 0

    for epoch in range(EPOCHS):

        model.train()
        for x,y in tqdm(train_loader,desc=f"Fold {fold} Epoch {epoch+1}"):
            x,y = x.to(device), y.to(device)

            optimizer.zero_grad()
            with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                out = model(x)
                loss = criterion(out,y)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

        scheduler.step()

        # Validation
        model.eval()
        preds,gts=[],[]

        with torch.no_grad():
            for x,y in val_loader:
                x,y = x.to(device),y.to(device)
                with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                    out=model(x)

                preds.extend(out.argmax(1).cpu().numpy())
                gts.extend(y.cpu().numpy())

        acc = accuracy_score(gts,preds)
        print(f"Epoch {epoch+1} Accuracy: {acc:.4f}")

        if acc>best:
            best=acc
            torch.save(model.state_dict(),f"best_fold{fold}.pth")

    fold_acc.append(best)

print("\nCV Mean Accuracy:", np.mean(fold_acc))

# ---------------- FINAL TEST EVALUATION ----------------
print("\nEvaluating best fold on Test set...")

best_fold = np.argmax(fold_acc)+1
model = EfficientNetSE(len(classes)).to(device)
model.load_state_dict(torch.load(f"best_fold{best_fold}.pth"))
model.eval()

test_ds = HistoDataset(test_files,test_labels,val_tf)
test_loader = DataLoader(test_ds,BATCH_SIZE,False,
                         num_workers=NUM_WORKERS,pin_memory=True)

preds,gts,probs=[],[],[]

with torch.no_grad():
    for x,y in test_loader:
        x,y = x.to(device),y.to(device)
        with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
            out=model(x)

        p = out.softmax(1)
        probs.extend(p.cpu().numpy())
        preds.extend(out.argmax(1).cpu().numpy())
        gts.extend(y.cpu().numpy())

acc  = accuracy_score(gts,preds)
bacc = balanced_accuracy_score(gts,preds)
f1m  = f1_score(gts,preds,average="macro")
auc  = roc_auc_score(label_binarize(gts,classes=range(len(classes))),
                     probs,average="macro",multi_class="ovr")

print("\n========== FINAL TEST RESULTS ==========")
print(f"Accuracy: {acc:.4f}")
print(f"Balanced Accuracy: {bacc:.4f}")
print(f"Macro F1: {f1m:.4f}")
print(f"Macro AUC: {auc:.4f}")

cm = confusion_matrix(gts,preds)
plt.figure(figsize=(6,5))
sns.heatmap(cm,annot=True,fmt="d",
            xticklabels=classes,yticklabels=classes)
plt.title("Test Confusion Matrix")
plt.show()
```

```python
# ============================================================
# GRAD-CAM (POST-TRAINING EXPLAINABILITY - COLAB SAFE)
# ============================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# -------- SELECT FOLD --------
fold = 1   # change if needed

# -------- RECREATE VALIDATION SPLIT (CORRECT VARIABLES) --------
splits = list(skf.split(train_files, train_labels))
_, va = splits[fold-1]

val_ds = HistoDataset(train_files[va], train_labels[va], val_tf)

# -------- LOAD TRAINED MODEL --------
model = EfficientNetSE(len(classes)).to(device)
model.load_state_dict(torch.load(f"best_fold{fold}.pth", map_location=device))
model.eval()

# -------- GRAD-CAM CLASS --------
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.gradients = None
        self.activations = None

        target_layer.register_forward_hook(self.forward_hook)
        target_layer.register_full_backward_hook(self.backward_hook)

    def forward_hook(self, module, input, output):
        self.activations = output

    def backward_hook(self, module, grad_input, grad_output):
        self.gradients = grad_output[0]

    def generate(self, input_tensor, class_idx=None):
        self.model.zero_grad()

        output = self.model(input_tensor)

        if class_idx is None:
            class_idx = output.argmax(dim=1).item()

        loss = output[:, class_idx]
        loss.backward()

        grads = self.gradients[0].detach().cpu().numpy()
        acts = self.activations[0].detach().cpu().numpy()

        weights = np.mean(grads, axis=(1, 2))

        cam = np.zeros(acts.shape[1:], dtype=np.float32)

        for i, w in enumerate(weights):
            cam += w * acts[i]

        cam = np.maximum(cam, 0)

        cam = cv2.resize(cam, (224, 224))
        cam -= cam.min()
        cam /= (cam.max() + 1e-8)

        return cam

# -------- TARGET LAYER --------
# EfficientNet-B0 last feature block
target_layer = model.backbone.features[-1]
gradcam = GradCAM(model, target_layer)

# -------- PICK VALIDATION SAMPLE --------
sample_index = 10  # change this to explore
img, label = val_ds[sample_index]

input_tensor = img.unsqueeze(0).to(device)

# -------- GET PREDICTION (NO GRAD) --------
with torch.no_grad():
    output = model(input_tensor)
    pred_class = output.argmax(dim=1).item()

# -------- GENERATE CAM (WITH GRAD) --------
cam = gradcam.generate(input_tensor, class_idx=pred_class)

# -------- VISUALIZE --------
img_np = img.permute(1, 2, 0).cpu().numpy()
img_uint8 = (img_np * 255).astype(np.uint8)

heatmap = (cam * 255).astype(np.uint8)
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

overlay = cv2.addWeighted(img_uint8, 0.6, heatmap, 0.4, 0)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
plt.title(f"Grad-CAM | True: {classes[label]} | Pred: {classes[pred_class]}")
plt.axis("off")
plt.show()

# -------- SAVE IMAGE --------
cv2.imwrite(f"gradcam_fold{fold}.png", overlay)
print(f"Saved: gradcam_fold{fold}.png")
```

```python
# ============================================================
# AUTOMATED ABLATION STUDY (COLAB VERSION - FOLD 1 ONLY)
# ============================================================

import os, torch, random
import numpy as np
import pandas as pd
from glob import glob
from tqdm import tqdm
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torchvision.io import read_image
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score

# ---------------- REPRODUCIBILITY ----------------
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

# ---------------- DEVICE ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

BATCH_SIZE = 96
EPOCHS = 8
NUM_WORKERS = 2

# ---------------- DATA (TRAIN ONLY) ----------------
BASE = "/content"
TRAIN_DIR = os.path.join(BASE, "Train_Images")

files, labels_str = [], []

for cls in sorted(os.listdir(TRAIN_DIR)):
    p = os.path.join(TRAIN_DIR, cls)
    if os.path.isdir(p):
        imgs = glob(os.path.join(p, "*"))
        files.extend(imgs)
        labels_str.extend([cls] * len(imgs))

files = np.array(files)
classes = sorted(list(set(labels_str)))
cls_to_idx = {c:i for i,c in enumerate(classes)}
labels = np.array([cls_to_idx[i] for i in labels_str])

print("Classes:", classes)
print("Total images:", len(files))

# ---------------- DATASET ----------------
class HistoDataset(Dataset):
    def __init__(self, files, labels, tf):
        self.files = files
        self.labels = labels
        self.tf = tf

    def __len__(self):
        return len(self.files)

    def __getitem__(self, i):
        img = read_image(self.files[i]).float() / 255.0
        img = self.tf(img)
        return img, self.labels[i]

# ---------------- SE BLOCK ----------------
class SEBlock(nn.Module):
    def __init__(self, ch, r=16):
        super().__init__()
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(ch, ch//r),
            nn.ReLU(),
            nn.Linear(ch//r, ch),
            nn.Sigmoid()
        )

    def forward(self, x):
        b,c,_,_ = x.shape
        y = self.pool(x).view(b,c)
        y = self.fc(y).view(b,c,1,1)
        return x*y

# ---------------- MODEL FACTORY ----------------
def build_model(num_classes, USE_SE):
    backbone = models.efficientnet_b0(weights="IMAGENET1K_V1")
    se = SEBlock(1280) if USE_SE else nn.Identity()

    class EfficientNetAblation(nn.Module):
        def __init__(self):
            super().__init__()
            self.backbone = backbone
            self.se = se
            self.backbone.classifier[1] = nn.Linear(1280, num_classes)

        def forward(self, x):
            x = self.backbone.features(x)
            x = self.se(x)
            x = self.backbone.avgpool(x)
            x = torch.flatten(x,1)
            x = self.backbone.classifier(x)
            return x

    return EfficientNetAblation()

# ---------------- SPLIT (FOLD 1 ONLY) ----------------
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
tr, va = list(skf.split(files, labels))[0]

# ---------------- ABLATION CONFIGS ----------------
configs = [
    {"name":"Baseline", "USE_SE":False, "USE_AUG":False, "FREEZE":False},
    {"name":"+SE", "USE_SE":True, "USE_AUG":False, "FREEZE":False},
    {"name":"+AUG", "USE_SE":False, "USE_AUG":True, "FREEZE":False},
    {"name":"+SE+AUG", "USE_SE":True, "USE_AUG":True, "FREEZE":False},
    {"name":"+SE+AUG+Freeze", "USE_SE":True, "USE_AUG":True, "FREEZE":True},
]

results = []

# ============================================================
# RUN ALL CONFIGS
# ============================================================

for cfg in configs:

    print(f"\nRunning Config: {cfg['name']}")

    train_tf = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(20)
    ]) if cfg["USE_AUG"] else transforms.Compose([
        transforms.Resize((224,224))
    ])

    val_tf = transforms.Compose([
        transforms.Resize((224,224))
    ])

    train_ds = HistoDataset(files[tr], labels[tr], train_tf)
    val_ds   = HistoDataset(files[va], labels[va], val_tf)

    train_loader = DataLoader(train_ds, BATCH_SIZE, True,
                              num_workers=NUM_WORKERS,
                              pin_memory=True)

    val_loader = DataLoader(val_ds, BATCH_SIZE, False,
                            num_workers=NUM_WORKERS,
                            pin_memory=True)

    model = build_model(len(classes), cfg["USE_SE"]).to(device)

    if cfg["FREEZE"]:
        for p in model.backbone.features.parameters():
            p.requires_grad = False

    optimizer = optim.AdamW(model.parameters(), lr=3e-4)
    criterion = nn.CrossEntropyLoss()

    # ---- Train ----
    for epoch in range(EPOCHS):
        model.train()
        for x,y in train_loader:
            x,y = x.to(device), y.to(device)
            optimizer.zero_grad()

            with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                out = model(x)
                loss = criterion(out,y)

            loss.backward()
            optimizer.step()

    # ---- Validate ----
    model.eval()
    preds, gts = [], []

    with torch.no_grad():
        for x,y in val_loader:
            x = x.to(device)
            out = model(x)
            preds.extend(out.argmax(1).cpu().numpy())
            gts.extend(y.numpy())

    acc  = accuracy_score(gts,preds)
    bacc = balanced_accuracy_score(gts,preds)
    f1m  = f1_score(gts,preds,average="macro")

    results.append([cfg["name"], acc, bacc, f1m])

# ============================================================
# FINAL COMPARISON TABLE
# ============================================================

results_df = pd.DataFrame(results,
                          columns=["Configuration",
                                   "Accuracy",
                                   "Balanced_Accuracy",
                                   "Macro_F1"])

print("\n================ ABLATION COMPARISON ================\n")
print(results_df.to_string(index=False))
print("\n=====================================================\n")
```

```python
# ============================================================
# FINAL BACKBONE COMPARISON (COLAB SAFE)
# 5-Fold Cross Validation
# ============================================================

import torch, os
import numpy as np
import pandas as pd
from tqdm import tqdm
import torch.nn as nn
import torch.optim as optim
from torchvision import models
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

EPOCHS = 5
NUM_WORKERS = 2

# ------------------------------------------------------------
# MODEL FACTORY
# ------------------------------------------------------------

def get_model(name, num_classes):

    if name == "resnet50":
        m = models.resnet50(weights="IMAGENET1K_V1")
        m.fc = nn.Linear(m.fc.in_features, num_classes)
        return m

    if name == "densenet121":
        m = models.densenet121(weights="IMAGENET1K_V1")
        m.classifier = nn.Linear(m.classifier.in_features, num_classes)
        return m

    if name == "mobilenetv3":
        m = models.mobilenet_v3_large(weights="IMAGENET1K_V1")
        m.classifier[3] = nn.Linear(m.classifier[3].in_features, num_classes)
        return m

    if name == "efficientnet":
        m = models.efficientnet_b0(weights="IMAGENET1K_V1")
        m.classifier[1] = nn.Linear(m.classifier[1].in_features, num_classes)
        return m

    if name == "efficientnet_se":

        class SEBlock(nn.Module):
            def __init__(self, ch, r=16):
                super().__init__()
                self.pool = nn.AdaptiveAvgPool2d(1)
                self.fc = nn.Sequential(
                    nn.Linear(ch, ch//r),
                    nn.ReLU(),
                    nn.Linear(ch//r, ch),
                    nn.Sigmoid()
                )

            def forward(self,x):
                b,c,_,_ = x.shape
                y = self.pool(x).view(b,c)
                y = self.fc(y).view(b,c,1,1)
                return x*y

        class EfficientNetSE(nn.Module):
            def __init__(self,num_classes):
                super().__init__()
                self.backbone = models.efficientnet_b0(weights="IMAGENET1K_V1")
                self.se = SEBlock(1280)
                self.backbone.classifier[1] = nn.Linear(1280,num_classes)

            def forward(self,x):
                x = self.backbone.features(x)
                x = self.se(x)
                x = self.backbone.avgpool(x)
                x = torch.flatten(x,1)
                x = self.backbone.classifier(x)
                return x

        return EfficientNetSE(num_classes)

# ------------------------------------------------------------
# MODELS TO COMPARE
# ------------------------------------------------------------

model_list = [
    "resnet50",
    "densenet121",
    "mobilenetv3",
    "efficientnet",
    "efficientnet_se"
]

comparison_results = []
total_steps = len(model_list) * 5
global_bar = tqdm(total=total_steps, desc="Overall Progress")

# ------------------------------------------------------------
# RUN COMPARISON
# ------------------------------------------------------------

for model_name in model_list:

    print("\n" + "="*60)
    print(f"RUNNING MODEL: {model_name.upper()}")
    print("="*60)

    fold_acc, fold_bacc, fold_f1, fold_auc = [],[],[],[]

    for fold,(tr,va) in enumerate(skf.split(train_files,train_labels),1):

        train_ds = HistoDataset(train_files[tr],train_labels[tr],train_tf)
        val_ds   = HistoDataset(train_files[va],train_labels[va],val_tf)

        train_loader = DataLoader(train_ds,BATCH_SIZE,True,
                                  num_workers=NUM_WORKERS,
                                  pin_memory=True)

        val_loader   = DataLoader(val_ds,BATCH_SIZE,False,
                                  num_workers=NUM_WORKERS,
                                  pin_memory=True)

        model = get_model(model_name,len(classes)).to(device)

        optimizer = optim.AdamW(model.parameters(),lr=3e-4)
        scaler = torch.cuda.amp.GradScaler(enabled=(device.type=="cuda"))
        criterion = nn.CrossEntropyLoss()

        # -------- TRAIN --------
        for epoch in range(EPOCHS):
            model.train()
            for x,y in train_loader:
                x,y = x.to(device), y.to(device)
                optimizer.zero_grad()

                with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                    out = model(x)
                    loss = criterion(out,y)

                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()

        # -------- VALIDATION --------
        model.eval()
        preds,gts,probs=[],[],[]

        with torch.no_grad():
            for x,y in val_loader:
                x = x.to(device)

                with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                    out = model(x)

                p = torch.softmax(out,1).cpu().numpy()
                probs.extend(p)
                preds.extend(out.argmax(1).cpu().numpy())
                gts.extend(y.numpy())

        acc  = accuracy_score(gts,preds)
        bacc = balanced_accuracy_score(gts,preds)
        f1m  = f1_score(gts,preds,average="macro")
        aucv = roc_auc_score(
            label_binarize(gts,classes=range(len(classes))),
            probs,
            average="macro",
            multi_class="ovr"
        )

        fold_acc.append(acc)
        fold_bacc.append(bacc)
        fold_f1.append(f1m)
        fold_auc.append(aucv)

        global_bar.update(1)

    comparison_results.append([
        model_name,
        np.mean(fold_acc),
        np.mean(fold_bacc),
        np.mean(fold_f1),
        np.mean(fold_auc)
    ])

global_bar.close()

# ------------------------------------------------------------
# FINAL TABLE
# ------------------------------------------------------------

comparison_df = pd.DataFrame(
    comparison_results,
    columns=["Model","Accuracy","Balanced_Accuracy","Macro_F1","Macro_AUC"]
)

print("\n================ FINAL MODEL COMPARISON ================\n")
print(comparison_df.sort_values("Macro_AUC", ascending=False).to_string(index=False))
print("\n========================================================\n")
```
